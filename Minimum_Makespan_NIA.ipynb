```python
# import libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
import os
data_collect = []
data_collect.append(pd.read_csv("Data1.csv"))
data_collect.append(pd.read_csv("Data2.csv")) 
data_collect.append(pd.read_csv("Data3.csv")) 
data_collect.append(pd.read_csv("Data4.csv")) 

data = pd.concat(data_collect, ignore_index=True)
```

Benchmark1 - Population_Size

```python
sns.violinplot(data = data, 
               x = "Changing Parameter", y='Value', 
               hue = "Algorithm Type", split = True, inner = "box", palette = ['b','r'])
plt.legend(loc='upper center')
plt.tight_layout()
```

```python
sns.boxplot(data = data, 
               x = "Changing Parameter", y='Time per Iteration', 
               hue = "Algorithm Type", palette = ['b','r'])
plt.legend(loc='upper center')
plt.tight_layout()
```
